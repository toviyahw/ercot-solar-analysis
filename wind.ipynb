### Temperature Data

```python
wind_temp_df = pd.read_csv('/Users/tovi/Downloads/era5_by_region.csv')
wind_temp_df.head()
```

```python
columns_to_drop = ['number', 'number.1', 'number.2', 'number.3']
wind_df_dropped = wind_temp_df.drop(columns=columns_to_drop, axis=1)
```

```python
wind_df_dropped.head()
```

```python
# List of temperature columns to convert
temp_columns = ['temp_2m_south', 'temp_2m_north', 'temp_2m_east', 'temp_2m_west']

# Convert from kelvin to celsius
wind_df_dropped[temp_columns] = wind_df_dropped[temp_columns] - 273.15

wind_df_dropped.head()
```

```python
# converting from celcius to fahrenheit
wind_df_dropped[temp_columns] = (wind_df_dropped[temp_columns] * 9/5) + 32
wind_df_dropped.head()
```

```python
wind_df_dropped.info()
```

```python
# saving csv to be joined with energy generation and hourly load data files using SQL
wind_df_dropped.to_csv('/Users/tovi/Downloads/wind_temps_converted.csv', index=False)
```

```python
"""
CREATE TABLE wind_temp_data (time TIMESTAMP NOT NULL,
							  temp_2m_south DECIMAL(5,2),
							  wind_u_south DECIMAL(14,10),
							  wind_v_south DECIMAL(14,10),
							  temp_2m_north DECIMAL (5,2),
							  wind_u_north DECIMAL(14,10),
							  wind_v_north DECIMAL(14,10),
							  temp_2m_east DECIMAL (5,2),
							  wind_u_east DECIMAL(14,10),
							  wind_v_east DECIMAL(14,10),
							  temp_2m_west DECIMAL (5,2),
							  wind_u_west DECIMAL(14,10),
							  wind_v_west DECIMAL(14,10)
							  );
"""
```

Returns 26280 rows as expected. DECIMAL data type precision and scale set to (14,10) to contain large windspeed float values. For temperature values it was set to (5,2) so that all values are rounded to 2 decimal places.

## Combining CSV Files Using SQL

```python
"""
SELECT wind_data.timestamp, temp_2m_south, wind_u_south, wind_v_south, temp_2m_north, wind_u_north, wind_v_north, temp_2m_east, wind_u_east,  wind_v_east, temp_2m_west, wind_u_west, wind_v_west, coast_load, east_load, farwest_load, north_load, northcentral_load, south_load, southcentral_load, west_load, system_load, wind_system, wind_coast, wind_south, wind_west, wind_north  
FROM wind_data
LEFT JOIN wind_temp_data
ON wind_data.timestamp = wind_temp_data.time
LEFT JOIN load_data
ON wind_data.timestamp = load_data.timestamp
ORDER BY wind_data.timestamp;
"""
```

This join returns all records in the wind_data table alongside the corresponding temperature and load values for matching timestamps from the temperature and load tables. Returns 25466 rows (check for duplicates in wind_data file).
